```python
# IMPORTS
import numpy as np
import pandas as pd

#Fin Data Sources
import yfinance as yf
import pandas_datareader as pdr

#Data viz
import plotly.graph_objs as go
import plotly.express as px

import time
from datetime import date
```

```python
end = date.today()
print(f'Year = {end.year}; month= {end.month}; day={end.day}')

start = date(year=end.year-10, month=end.month, day=end.day)
print(f'Period for indexes: {start} to {end} ')
```

```python
# INDEXES from Yahoo Finance
# Nifty index (XETRA - XETRA Delayed Price. Currency in EUR)
# WEB: https://finance.yahoo.com/quote/%5EGDAXI
nifty_daily = yf.download(tickers = "^NSEI", period = "max", interval = "1d")
```

```python
nifty_daily.tail()
```

```python
# normally 252 trading days
nifty_daily['adj_close_last_year'] = nifty_daily['Adj Close'].shift(252)
nifty_daily['yoy_growth'] = nifty_daily['Adj Close'] / nifty_daily['adj_close_last_year'] -1
```

```python
nifty_daily
```

```python
nifty_daily['Adj Close'].plot.line()
```
